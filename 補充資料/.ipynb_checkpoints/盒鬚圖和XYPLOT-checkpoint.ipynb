```python
import matplotlib.pyplot as plt
import numpy as np

# Data Scatter Plot
x= np.arange(2,7,1)
y=np.arange(75,105,5)
#print(x)
y1= [87,92,96,98,98]
y2 = [85,90,93,94,96]
y3 = [79,80,90,97,97]

dy1= [0.75,0.8,0.6,0.4,0.5]
dy2 =[0.5,0.4,0.6,0.4,0.25]
dy3 =[0.4,0.7,0.6,0.4,0.3]

z1 = np.polyfit(x, y1, 3)
z2 = np.polyfit(x, y2, 3)
z3 = np.polyfit(x, y3, 3)
#print(z)
p1 = np.poly1d(z1)
p2 = np.poly1d(z2)
p3 = np.poly1d(z3)
#print(p)

plt.plot(np.linspace(2,6),p1(np.linspace(2,6)),"#FFCC00")
plt.plot(np.linspace(2,6),p2(np.linspace(2,6)),"#CC0033")
plt.plot(np.linspace(2,6),p3(np.linspace(2,6)),"#CCCCCC")

plt.xticks(x)
plt.yticks(y)
plt.errorbar(x,y1,yerr=dy1,fmt='o',ecolor='#FFCC00',color='#FFCC00',elinewidth=3,capsize=4,label='Soybean')
plt.errorbar(x,y2,yerr=dy2,fmt='o',ecolor='#CC0033',color='#CC0033',elinewidth=3,capsize=4,label='Red bean')
plt.errorbar(x,y3,yerr=dy3,fmt='o',ecolor='#CCCCCC',color='#CCCCCC',elinewidth=3,capsize=4,label='White bean')
plt.xlabel("Deph [# of layers]")
plt.ylabel("Accuracy [%]")
plt.legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(777)
ydata = [np.random.normal(88, 3, size=100),np.random.normal(92, 3, size=100),np.random.normal(94, 2, size=100),
        np.random.normal(96, 1.5, size=100),np.random.normal(96, 1.5, size=100)]

    
labels = ['2', '3', '4','5','6']

bplot = plt.boxplot(ydata, patch_artist=True, labels=labels)  # 设置箱型图可填充


colors = ['#CCFF66','pink', 'lightblue', '#CCCCFF','#FFCC99']
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)  # 为不同的箱型图填充不同的颜色

x= np.arange(0,6,1)
y_test1= [99,99,99,99,99,99]
z_past1 = np.polyfit(x, y_test1, 1)
p_past1 = np.poly1d(z_past1)
plt.plot(np.linspace(0,6),p_past1(np.linspace(0,6)),"#FFCC00",linestyle="--",label="Soybean - Best from Larese et al.(2014)")

y=np.arange(80,105,5)
plt.xlabel("Deph [# of layers]")
plt.ylabel("Accuracy [%]")
plt.legend(loc='best')
plt.yticks(y)
plt.show()
```
